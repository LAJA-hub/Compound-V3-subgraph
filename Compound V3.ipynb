```python
import pandas as pd
import random
import matplotlib.pyplot as plt

# Mock Data for Positions
positions_data = [
    {"positions_id": "pos_1", "positions_account_address": "0x123"},
    {"positions_id": "pos_2", "positions_account_address": "0x456"},
    {"positions_id": "pos_3", "positions_account_address": "0x789"},
]

# Create a DataFrame for positions
positions_df = pd.DataFrame(positions_data)

# Function to simulate fetching snapshots for a position
def fetch_position_snapshots(position_id):
    # Simulate snapshot data
    snapshot_data = [
        {"timestamp": i, "accounting": random.randint(100, 1000)}
        for i in range(1, 11)  # Simulate 10 snapshots
    ]
    return pd.DataFrame(snapshot_data)

# Function to plot snapshots
def plot_all(dataframes, x_columns, title):
    for df, x_col in zip(dataframes, x_columns):
        plt.plot(df[x_col], df["accounting"], label=x_col)
    plt.title(title)
    plt.xlabel("Timestamp")
    plt.ylabel("Accounting Value")
    plt.legend()
    plt.show()

# Iterate through positions and fetch their snapshots
for index, row in positions_df.iterrows():
    position_address = row["positions_account_address"]
    position_id = row["positions_id"]

    # Fetch snapshots for the position
    snapshots_df = fetch_position_snapshots(position_id)

    # Display snapshots
    print(f"Snapshots for Position {position_address} (ID: {position_id})")
    print(snapshots_df)

    # Plot the snapshots
    plot_all([snapshots_df], ["timestamp"], f"{position_address} - Position Accounting")
```
